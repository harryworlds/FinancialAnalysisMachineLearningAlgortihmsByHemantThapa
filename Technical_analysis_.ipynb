Technical analysis in the stock market is a method of evaluating securities by analyzing statistical trends derived from market data, primarily price and volume. Technical analysts believe that the historical performance of stocks and markets are indications of future performance.


Trends: Technical analysts believe that trends, both short-term and long-term, exist in the stock market and that these trends can be used to make investment decisions.

Support and resistance levels: These are the levels at which a stock's price experiences difficulty in moving beyond. Support levels are typically seen as levels that a stock's price will not fall below, while resistance levels are seen as levels that a stock's price will not rise above.

Chart patterns: Technical analysts use charts to identify patterns, such as head and shoulders or bullish and bearish reversal patterns, that can indicate a potential trend reversal.

Indicators: Technical analysts use mathematical calculations based on the price and/or volume of a stock to generate indicators, such as moving averages, momentum indicators, and oscillators, that can provide insight into the strength and direction of a stock's trend.

It is important to note that technical analysis is not a perfect method and should not be used as the sole basis for investment decisions. It is also not a universally accepted method and there are many who believe that it has limited usefulness in the stock market. It is best used as a complementary tool in conjunction with other forms of analysis, such as fundamental analysis.

### 2. TECHNICAL ANALYSIS

```python
def technical_analysis(ticker):
    stock = yf.download(ticker.upper(), start="2013-02-10", end="2023-02-10")
    
    #Open and closed price
    plt.figure(figsize=(15,6), facecolor="black")
    plt.title(f"{ticker.upper()} - Open v/s Close ", fontsize=16, color='white')
    plt.plot(stock.index, stock.Open, color="green", label="Open Price")
    plt.plot(stock.index, stock.Close, color="red", label="Close Price")
    plt.ylabel("Price ($)")
    plt.xlabel("Date")
    plt.xticks(fontsize=12, rotation=45, color='white')
    plt.yticks(fontsize=12, color='white')
    plt.grid(True, color='grey')
    plt.show()

    #Open Price
    plt.figure(figsize=(15,6), facecolor="black")
    plt.title(f"{ticker.upper()} - OPEN ", fontsize=16, color='white')
    plt.plot(stock.index, stock.Open, color="green")
    plt.bar(stock.index, stock.Open, color="green", label="Open Price")
    plt.ylabel("Price ($)",fontsize=14, color='white')
    plt.xlabel("Date",fontsize=14, color='white')
    plt.xticks(fontsize=12, rotation=45, color='white')
    plt.yticks(fontsize=12, color='white')
    plt.grid(True, color='grey')
    plt.show()
    
    #Close Price
    plt.figure(figsize=(15,6), facecolor="black")
    plt.title(f"{ticker.upper()} - CLOSE ", fontsize=16, color='white')
    plt.plot(stock.index, stock.Close, color="red")
    plt.bar(stock.index, stock.Close, color="red", label="Close Price")
    plt.ylabel("Price ($)",fontsize=14, color='white')
    plt.xlabel("Date",fontsize=14, color='white')
    plt.xticks(fontsize=12, rotation=45, color='white')
    plt.yticks(fontsize=12, color='white')
    plt.grid(True, color='grey')
    plt.show()
    
    #High and low
    plt.figure(figsize=(15,6), facecolor="black")
    plt.title(f"{ticker.upper()} - High v/s Low ", fontsize=16, color='white')
    plt.plot(stock.index, stock.High, color="Blue", label="High Price")
    plt.plot(stock.index, stock.Low, color="red", label="Low Price")
    plt.ylabel("Price ($)",fontsize=14, color='white')
    plt.xlabel("Date",fontsize=14, color='white')
    plt.xticks(fontsize=12, rotation=45, color='white')
    plt.yticks(fontsize=12, color='white')
    plt.grid(True, color='grey')
    plt.show()
```

```python
def overlay(ticker1, ticker2):
    
    intial_date="2010-02-10"
    final_date="2023-02-10"
    stock_list = [ticker1, ticker2]
    stock = []
    for i in stock_list:
        data = yf.download(i.upper(), start=intial_date, end=final_date)
        stock.append(data)
    
    #Open and closed price
    plt.figure(figsize=(15,6), facecolor="black")
    plt.title(f"{ticker.upper()} - Open v/s Close ", fontsize=16, color='white')
    plt.plot(stock[0].index, stock[0].Open, color="green", label="Open Price")
    plt.plot(stock[1].index, stock[1].Open, color="red", label="Open Price")
    plt.ylabel("Price ($)")
    plt.xlabel("Date")
    plt.xticks(fontsize=12, rotation=45, color='white')
    plt.yticks(fontsize=12, color='white')
    plt.grid(True, color='grey')
    plt.show()
    
```

### 3. Linear Regression Model

#### Linear Regression: Simple linear regression models can be used to predict stock prices if the relationship between the independent and dependent variables is linear.

```python
from sklearn.linear_model import LinearRegression
```

```python
snp500 = yf.Ticker("^GSPC")
snp500_history = snp500.history(period="10y")

microsoft = yf.Ticker("MSFT")
microsoft_history = microsoft.history(period="10y")

gs = yf.Ticker("GS")
gs_history = gs.history(period="10y")
```

```python
snp500_history[:5]
```

```python
#Calling unique columns
print(snp500_history.columns)
print(microsoft_history.columns)
print(gs_history.columns)
```

```python
#Converting into lower case
snp500_history.columns = snp500_history.columns.str.lower()
microsoft_history.columns = microsoft_history.columns.str.lower()
gs_history.columns = gs_history.columns.str.lower()
```

```python
#Calling unique columns
print(snp500_history.columns)
print(microsoft_history.columns)
print(gs_history.columns)
```

```python
dataset = {"snp500":snp500_history['open'],
            "microsoft":microsoft_history['open'],
           "goldman_sachs":gs_history['open'],
          }
```

```python
df = pd.DataFrame(dataset)
df.tail(10)
```

```python
Model = LinearRegression()
```

```python
y = df["goldman_sachs"]
y = y.values.reshape(-1,1)
y
```

```python
X = df["snp500"]
X = X.values.reshape(-1,1)
X[:10]
```

```python
Model.fit(X,y)
```

```python
print(f'Coefficients: {round(Model.coef_[0][0],4)}')
print(f'Interception: {round(Model.intercept_[0],4)}')
```

1. A positive sign indicates that as the predictor variable increases, the response variable also increases.

2. A negative sign indicates that as the predictor variable increases, the response variable decreases.

```python
predict_y = Model.predict(X)
predict_y[:10]
```

```python
print(f"\nModel Accuracy: {round(Model.score(X,y)*100,4)}%\n")
plt.figure(figsize=(15,6), facecolor="grey")
sns.scatterplot(data=df, x=df['snp500'], y=df['goldman_sachs'], color='black', label="SN500 & Goldman Sachs")
plt.scatter(df["snp500"], predict_y, linewidth=0.5, color="red", marker='+', label="Prediction")
plt.title("SKlearn - Linear Regression Model", fontsize=16, color="white")
plt.legend(fontsize=12, facecolor='grey', edgecolor='white')
plt.ylabel("Goldman Sachs", fontsize=15, color="white")
plt.xlabel("SNP500", fontsize=15, color="white")
plt.yticks(fontsize=12, color="white")
plt.grid(True, color="grey")
plt.show()
```

```python
print(f"\nModel Accuracy: {(Model.score(X,y))}\n")
```
